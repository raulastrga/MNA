# 4. Base de conocimiento (RAG)

## Retrieval-Augmented Generation (RAG)

A diferencia de un sistema RAG (Retrieval-Augmented Generation) tradicional, que sigue un flujo lineal de "pregunta-búsqueda-respuesta", este proyecto implementa una variante agéntica. En este enfoque, un Orquestador toma decisiones dinámicas sobre qué herramientas utilizar, permitiendo que el sistema trascienda las limitaciones de la recuperación estática mediante el razonamiento complejo (Singh et al., 2025).

La arquitectura diseñada se fundamenta en la fragmentación inteligente (Chunking) y el procesamiento de documentos Markdown. Para asegurar que los agentes de Becas y Carreras mantengan el contexto, se cargaron los archivos mediante TextLoader y se dividieron en fragmentos de 900 a 1000 caracteres con un traslape (overlap) de 150. Mientras LangChain utiliza el RecursiveCharacterTextSplitter para esta tarea, la arquitectura de Microsoft Agent Framework (MAF) emplea lógica personalizada basada en expresiones regulares para respetar la jerarquía del documento.

**Estrategias de Recuperación y Búsqueda**

La implementación permite comparar dos filosofías de recuperación:

- Recuperación Semántica (LangChain): Utiliza Vector Embeddings de alta dimensión mediante el modelo sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2. Estos vectores se almacenan en una base de datos FAISS, permitiendo una búsqueda basada en conceptos y significados. Como señalan Gun y Oksanen (2025), este método convierte la consulta en un vector para recuperar fragmentos basados en su distancia matemática en un espacio continuo.

- Recuperación Léxica (MAF): Utiliza un enfoque basado en frecuencias de términos (TF-IDF) y matrices en memoria. Aunque la búsqueda semántica es ideal para capturar relaciones complejas (Wampler et al., 2025), esta arquitectura demuestra que el análisis léxico robusto es altamente efectivo para extraer datos técnicos específicos y precisos del corpus.

Esta dualidad en el diseño permite que los agentes especialistas extraigan información relevante de manera eficiente, optimizando la precisión del sistema a medida que el corpus de conocimientos escala (Sawarkar et al., 2024).

### Markdown

Markdown es un lenguaje de marcado ligero que permite aplicar formato a un texto utilizando símbolos sencillos de leer tanto para humanos como para máquinas. A diferencia de los procesadores de texto convencionales, Markdown estructura la información mediante etiquetas de jerarquía (como encabezados, listas y tablas) sin añadir metadatos ocultos ni códigos complejos. Esto lo convierte en un formato de "texto puro" altamente eficiente, que conserva la organización lógica del contenido mientras mantiene un peso mínimo de archivo.

Se eligió Markdown como la estructura central de la base de conocimiento debido a su capacidad superior para ser procesado por modelos de lenguaje en arquitecturas RAG. Dado que la documentación original se encontraba en formato Microsoft Word (.docx) —un formato binario con exceso de ruido visual y estructural—, la conversión a Markdown permitió limpiar el contenido y preservar únicamente la semántica esencial. Esta decisión se alinea con las mejores prácticas actuales, ya que, como señala Conner (2025), la estructuración del contenido es un factor crítico para el éxito de los sistemas RAG, pues el uso de formatos que preserven la jerarquía técnica permite que el modelo recupere información más precisa y contextualmente relevante, reduciendo significativamente las alucinaciones del sistema.

## 5.1. Evaluación Automática de Calidad: Fidelidad (Faithfulness)

Un paso crítico e importante es evaluar la consistencia factual de las respuestas generadas para garantizar que el sistema no presente alucinaciones. Para automatizar este proceso a escala, se integró el framework Ragas (Retrieval Augmented Generation Assessment), empleando un enfoque de LLM-as-a-Judge (un modelo de lenguaje potente actuando como juez evaluador).

Específicamente, la función implementada calcula la métrica de Fidelidad (Faithfulness). Esta métrica cuantifica qué tan fiel es la respuesta generada con respecto al contexto recuperado de la base de conocimientos. El sistema penaliza al agente si su respuesta incluye afirmaciones inventadas, datos externos o detalles que no se pueden deducir estrictamente de los fragmentos en formato Markdown o PostgreSQL que se le proporcionaron.

De acuerdo con Es et al. (2023), creadores del framework Ragas, la evaluación automatizada de la fidelidad se logra extrayendo primero un conjunto de afirmaciones atómicas de la respuesta generada, para luego utilizar un modelo juez que verifica si cada una de estas afirmaciones puede inferirse lógicamente del contexto recuperado. Esto asegura una auditoría técnica confiable sobre la precisión de los agentes.

```python
from datasets import Dataset
from ragas import evaluate
from ragas.metrics import faithfulness

def evaluate_models(pregunta, judge_llm, retriever, respuesta_generada):
  # 1. Estructura para guardar los datos
  data_samples = {
      'question': [],
      'answer': [],
      'contexts': []
  }

  # 2. Se usa el retriever directamente para obtener el contexto
  docs_objetos = retriever.invoke(pregunta)

  # 3. Capturar el contexto automáticamente
  contextos_recuperados = []
  for doc in docs_objetos:
    contextos_recuperados.append(doc.page_content)

  # 4. Guardar en el diccionario
  data_samples['question'].append(pregunta)
  data_samples['answer'].append(respuesta_generada)
  data_samples['contexts'].append(contextos_recuperados)

  # 5. Crear el Dataset para Ragas
  dataset_ragas = Dataset.from_dict(data_samples)

  # 6. Ejecutar la evaluación de Faithfulness
  score = evaluate(
      dataset_ragas,
      metrics=[faithfulness],
      llm=judge_llm,
  )

  score_lista = score["faithfulness"]

  try:
      score_numerico = float(score_lista[0])
  except TypeError:
      score_numerico = float(score_lista)

  print('✅ Score Faithfulness:', score_numerico)

  return score_numerico
```

## 5.2. Langchain

LangChain es un framework de desarrollo de código abierto diseñado para facilitar la creación de aplicaciones que utilizan modelos de lenguaje de gran escala (LLMs). Su arquitectura se basa en el concepto de "cadenas" (chains), las cuales permiten secuenciar diferentes componentes como prompts, modelos y herramientas de recuperación de datos (RAG) para ejecutar tareas complejas de manera automatizada. En este proyecto, LangChain actúa como el pegamento estructural que conecta a los agentes con la base de conocimientos en Markdown y la base de datos de eventos, gestionando la memoria y el historial de la conversación para ofrecer respuestas coherentes y contextualizadas.

De acuerdo con la documentación oficial del ecosistema:

> "LangChain está diseñado para ayudar a los desarrolladores a construir aplicaciones que no solo realicen llamadas a un LLM a través de una API, sino que también sean conscientes de los datos (conectando el modelo con fuentes externas) y agénticas (permitiendo que el modelo interactúe con su entorno)."
(Chase, 2022).

```python
# =========================
# Imports / Dependencias
# =========================

# LangChain: wrappers y componentes base para prompts, parsing y modelos
from langchain_openai import ChatOpenAI
from langchain_huggingface import HuggingFaceEndpoint, ChatHuggingFace
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain.chat_models import init_chat_model

# RAG: carga de documentos, chunking y vector store
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS
from langchain_huggingface import HuggingFaceEndpointEmbeddings

# BD: conexión a PostgreSQL (eventos)
import psycopg2

# Cache: evita reinstanciar modelos con la misma configuración repetidamente
from functools import lru_cache

# ============================================================
# 1) Registry de modelos (normaliza múltiples proveedores)
# ============================================================
MODEL_REGISTRY = {
    # Azure Foundry Models (OpenAI v1 compatible)
    "deepseek_v3_2": {
        "provider": "azure_foundry_v1",
        "model": "DeepSeek-V3.2",
        "name": "DeepSeek-V3.2",
    },
    "gpt_5_chat": {
        "provider": "azure_foundry_v1",
        "model": "gpt-5-chat",
        "name": "gpt-5-chat",
    },
    "grok_3": {
        "provider": "azure_foundry_v1",
        "model": "grok-3",
        "name": "grok-3",
    },
    "llama_4_maverick": {
        "provider": "azure_foundry_v1",
        "model": "Llama-4-Maverick-17B-128E-Instruct-FP8",
        "name": "Llama-4-Maverick-17B-128E-Instruct-FP8",
    },
}

# ============================================================
# 2) Factoría de modelos (con cache)
# ============================================================
@lru_cache(maxsize=16)
def get_chat_model(model_key: str, temperature: float = 0.1, max_new_tokens: int = 2048):
    """
    Devuelve un Chat Model (Runnable) según el model_key.

    ¿Por qué cache?
    - En notebooks/servicios, reinstanciar clientes LLM en cada llamada
      es costoso (latencia, overhead, rate limits, etc.).
    - lru_cache evita reinstanciar mientras la combinación de parámetros
      sea la misma (model_key, temperature, max_new_tokens).

    Parámetros:
    - model_key: clave dentro de MODEL_REGISTRY.
    - temperature: control de aleatoriedad (bajo para tareas determinísticas).
    - max_new_tokens: tope de tokens generados (más alto => más costo/latencia).
    """
    if model_key not in MODEL_REGISTRY:
        raise ValueError(f"Modelo '{model_key}' no existe. Opciones: {list(MODEL_REGISTRY.keys())}")

    cfg = MODEL_REGISTRY[model_key]

    # Exponemos nombre del modelo seleccionado
    global modelo_seleccionado
    modelo_seleccionado = cfg["name"]

    # --- Provider: Azure OpenAI ---
    if cfg["provider"] == "azure_foundry_v1":
        base_url = os.getenv("FOUNDRY_OPENAI_BASE_URL")
        api_key = os.getenv("FOUNDRY_OPENAI_API_KEY")

        if not base_url or not api_key:
            raise ValueError("Faltan variables: FOUNDRY_OPENAI_BASE_URL y/o FOUNDRY_OPENAI_API_KEY")

        # LangChain indica que para Azure OpenAI v1 se use base_url con /openai/v1/.  [oai_citation:2‡docs.langchain.com](https://docs.langchain.com/oss/python/integrations/chat/openai?utm_source=chatgpt.com)
        return ChatOpenAI(
            model=cfg["model"],
            base_url=base_url,
            api_key=api_key,
            temperature=temperature,
            max_tokens=max_new_tokens,  # en ChatOpenAI el parámetro estándar es max_tokens
        )

    # Fallback: provider no soportado
    raise ValueError(f"Provider no soportado: {cfg['provider']}")

# ============================================================
# 3) Conexión a Base de Datos (Eventos)
# ============================================================
def get_db_connection():
    """
    Crea una conexión a PostgreSQL.

    Nota importante:
    - Se asume que estas variables existen en el entorno:
      BD_HOST, BD_SCHEMA, BD_USER, BD_PASSWORD.
    """
    conn = psycopg2.connect(
        host=BD_HOST,
        database=BD_SCHEMA,
        user=BD_USER,
        password=BD_PASSWORD,
        port=5432
    )
    return conn

# ============================================================
# 4) RAG: Índices (Carreras / Becas)
# ============================================================
FILE_KB_PATH = "kb_total.md"

def _build_kb_retriever_from_md():
    """
    Construye un retriever (FAISS) a partir de un archivo Markdown.
    Pipeline típico:
    1) Load documento
    2) Split en chunks (para granularidad y recall)
    3) Embeddings (modelo denso)
    4) Indexación en FAISS
    5) Retriever (k documentos top)
    """
    # 1) Cargar markdown
    loader = TextLoader(FILE_KB_PATH, encoding="utf-8")
    docs = loader.load()

    # 2) Dividir en chunks. Separators favorecen cortes por encabezados y listas.
    splitter = RecursiveCharacterTextSplitter(
        chunk_size=1000,
        chunk_overlap=150,
        separators=["\n## ", "\n### ", "\n#### ", "\n- ", "\n", " "]
    )
    chunks = splitter.split_documents(docs)

    # 3) Embeddings multilingües (útil si KB contiene español/inglés)
    embeddings = HuggingFaceEndpointEmbeddings(
        repo_id="sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2",
        huggingfacehub_api_token=os.environ["HUGGINGFACEHUB_API_TOKEN"]
    )

    # 4) Indexación FAISS en memoria
    vectorstore = FAISS.from_documents(chunks, embeddings)

    # 5) Retriever con top-k
    retriever = vectorstore.as_retriever(search_kwargs={"k": 4})
    return retriever

# --- Construcción de retrievers con tolerancia a fallos ---
# Si falla la construcción, el sistema seguirá vivo y responderá con "FALTA_INFORMACION".
try:
    carreras_retriever = _build_kb_retriever_from_md()
    print("✅ RAG de carreras listo (Markdown indexado).")
except Exception as e:
    carreras_retriever = None
    print(f"⚠️ No se pudo construir el RAG de carreras desde Markdown: {e}")

try:
    becas_retriever = _build_kb_retriever_from_md()
    print("✅ RAG de becas listo (misma base Markdown indexada).")
except Exception as e:
    becas_retriever = None
    print(f"⚠️ No se pudo construir el RAG de becas desde Markdown: {e}")


# ============================================================
# 5) Agentes de apoyo (helpers)
# ============================================================
def extraer_campus(consulta, chat_model):
    """
    Extrae el campus/ciudad mencionado en la consulta usando el LLM.

    Diseño:
    - Un extractor dedicado reduce reglas manuales/frágiles.
    - Respuesta restringida (solo nombre o NONE) para facilitar lógica downstream.

    Retorna:
    - Nombre del campus/ciudad (string) o "NONE".
    """
    prompt = ChatPromptTemplate.from_messages([
        ("system", """Eres un extractor de entidades. Tu único trabajo es identificar si el usuario menciona un nombre de Ciudad o Campus en su texto.
        - Si encuentras uno, responde SOLO con el nombre (ej: 'Monterrey', 'Santa Fe', 'Guadalajara').
        - Si no encuentras ninguna ubicación, responde la palabra 'NONE'.
        - No añadas puntuación ni palabras extra."""),
        ("user", f"Pregunta: {consulta}")
    ])
    chain = prompt | chat_model
    return chain.invoke({})

def ejecutar_sintetizador(pregunta_original, respuesta_agente, chat_model):
    """
    Sintetizador / post-procesador:
    - Toma la salida "técnica" (RAG/DB) y la convierte en una respuesta final
      amigable, breve y alineada a la pregunta original.
    - Sirve como capa anti-alucinación: obliga a hablar solo con lo provisto.

    Importante:
    - Este agente NO debe introducir hechos nuevos.
    - Debe omitir detalles internos (errores técnicos, BD, tokens, etc.).
    """
    prompt_sintetizador = ChatPromptTemplate.from_messages([
        ("system", """Eres un sintetizador de información experto. Actúa como un asistente de atención al usuario **amable, empático y muy eficiente**.

        Tu tarea es comunicar la información técnica al usuario siguiendo estas REGLAS DE ORO:

        1. **BREVEDAD RADICAL:** Tu respuesta debe ser breve pero no descartes información importante para el usuario. Sé directo pero suave.
        2. **TONO:** Usa un lenguaje cercano y servicial (Ej: "Con gusto te comparto...", "Claro, aquí tienes...").
        3. **COHERENCIA:** Responde directamente a la pregunta del usuario usando SOLO la información provista.
        4. **SIN RELLENO:** Elimina saludos largos o despedidas innecesarias. Ve al grano con amabilidad.
        5. NO inventes información extra (no alucinar).
        6. NO adivines fechas o requisitos que no estén en la respuesta técnica.
        7. Si la respuesta técnica dice que falta información, menciónalo tal cual, no menciones tecnicismos acerca de bases de datos, conexiones o errores.
        8. Asegúrate de responder directamente a lo que el usuario preguntó originalmente.
        9. Si se trata de eventos, muestra una lista completa de las opciones con un formato entendible y amigable para el usuario.
        """),
        ("user", f"Pregunta del usuario: {pregunta_original}\n\nInformación técnica cruda: {respuesta_agente}\n\nRespuesta final amable y breve:")
    ])

    chain = prompt_sintetizador | chat_model
    return chain.invoke({})

# ============================================================
# 6) Agentes especializados (especialistas por dominio)
# ============================================================
def ejecutar_becas(consulta, chat_model):
    """
    Especialista BECAS:
    - Implementa RAG sobre kb_total.md.
    - Responde SOLO con lo recuperado del contexto.
    - Si falta información, responde con un prefijo controlado (FALTA_INFORMACION).

    Convención de estados:
    - FALTA_INFORMACION: el sistema no puede responder sin más datos o sin KB.
    - SIN_RESULTADOS: el retriever no encontró fragmentos relevantes.
    - ERROR_SISTEMA: fallas técnicas (para debugging; el sintetizador debe suavizarlo).
    """
    if becas_retriever is None:
        return "FALTA_INFORMACION: No está disponible la base de conocimiento de becas (archivo Markdown) para consultar en este momento."

    # Enriquecemos la consulta para sesgar el retrieval al dominio de becas
    query_rag = f"BECAS / APOYOS ECONÓMICOS / FINANCIAMIENTO: {consulta}"

    try:
        docs = becas_retriever.invoke(query_rag)
    except Exception as e:
        return f"ERROR_SISTEMA: No se pudo consultar el índice de becas: {str(e)}"

    if not docs:
        return "SIN_RESULTADOS: No encontré información de becas en la base. Pide al usuario que especifique el tipo de beca (académica, socioeconómica, deportiva, etc.) o el campus/periodo."

    # Unimos el contenido recuperado (contexto) para el prompt
    contexto = "\n\n".join([d.page_content for d in docs])

    # Prompt con reglas estrictas para minimizar alucinaciones
    prompt = ChatPromptTemplate.from_messages([
        ("system", """Eres un asistente experto en BECAS y APOYOS ECONÓMICOS del Tecnológico de Monterrey.

REGLAS ESTRICTAS:
- Responde ÚNICAMENTE usando el CONTEXTO proporcionado (extraído del archivo Markdown).
- SOLO responde sobre becas/apoyos/financiamiento. Si el contexto recuperado trata de carreras u otro tema, ignóralo.
- Si el contexto NO contiene la respuesta, responde iniciando con: "FALTA_INFORMACION:" e indica qué dato falta (tipo de beca, campus, nivel, periodo, etc.).
- NO inventes montos, porcentajes, fechas, requisitos o procesos.
- Si hay varias becas/opciones en el contexto, lista las opciones en viñetas y pide al usuario elegir.

FORMATO:
- Breve y claro.
- Usa viñetas para requisitos/pasos si aplica.
"""),
        ("user", "Pregunta: {pregunta}\n\nCONTEXTO (Markdown):\n{contexto}\n\nRespuesta:")
    ])

    chain = prompt | chat_model
    return chain.invoke({"pregunta": consulta, "contexto": contexto})

def ejecutar_carreras(consulta, chat_model):
    """
    Especialista CARRERAS:
    - Implementa RAG sobre kb_total.md.
    - Responde SOLO con lo recuperado del contexto.
    - Prefijos controlados para estados (misma convención que becas).
    """
    if carreras_retriever is None:
        return "FALTA_INFORMACION: No está disponible la base de conocimiento de carreras (archivo Markdown) para consultar en este momento."

    # 1) Recuperación
    try:
        docs = carreras_retriever.invoke(consulta)
    except Exception as e:
        return f"ERROR_SISTEMA: No se pudo consultar el índice de carreras: {str(e)}"

    if not docs:
        return "SIN_RESULTADOS: No encontré información relacionada en la base de carreras. Pide al usuario que especifique el nombre exacto de la carrera o el área (Ej. Ingeniería, Negocios, Salud)."

    contexto = "\n\n".join([d.page_content for d in docs])

    # 2) Prompt controlado para evitar alucinaciones
    prompt = ChatPromptTemplate.from_messages([
        ("system", """Eres un asistente experto en programas académicos/carreras del Tecnológico de Monterrey.

REGLAS ESTRICTAS:
- Responde ÚNICAMENTE usando el CONTEXTO proporcionado (extraído del archivo Markdown).
- Si el contexto no contiene la respuesta, responde iniciando con: "FALTA_INFORMACION:" y pide el dato que falte.
- No inventes requisitos, costos, sedes, fechas, duración, ni planes de estudio.
- Si hay varias opciones en el contexto, lista las opciones y pide al usuario que elija.
- Si el usuario pide comparación, compara SOLO lo que esté explícito en el contexto.

FORMATO:
- Breve, claro.
- Usa viñetas si ayuda.
"""),
        ("user", "Pregunta: {pregunta}\n\nCONTEXTO (Markdown):\n{contexto}\n\nRespuesta:")
    ])

    chain = prompt | chat_model
    return chain.invoke({"pregunta": consulta, "contexto": contexto})

def ejecutar_eventos(consulta, chat_model):
    """
    Especialista EVENTOS (DB-driven):
    Flujo:
    1) Extraer campus con LLM (entity extraction).
    2) Consultar PostgreSQL con coincidencia parcial (ILIKE) por campus/nombre.
    3) Filtrar por eventos futuros (>= CURRENT_DATE).
    4) Normalizar texto/fechas con un LLM "limpiador" (determinístico).

    Nota:
    - Este agente devuelve información "cruda/estructurada" que luego el
      sintetizador convierte en respuesta final al usuario.
    """
    # 1) Extraer campus desde la consulta
    resp_campus = extraer_campus(consulta, chat_model)
    campus_detectado = resp_campus.content.strip()
    info_eventos = ""

    print(f"🤖 [LOG] Campus detectado: {campus_detectado}")

     # Si no hay campus, pedimos más contexto al usuario
    if campus_detectado == "NONE":
        return "FALTA_INFORMACION: No se detectó un campus en la consulta. Se requiere que el usuario especifique una ciudad o campus para buscar sesiones informativas."

    try:
        # 2) Abrir conexión a la BD
        conn = get_db_connection()
        cur = conn.cursor()

        # Consulta:
        # - ILIKE: búsqueda case-insensitive con coincidencia parcial
        # - CURRENT_DATE: solo eventos futuros / de hoy
        # - LIMIT 3: control de volumen de salida (ajustable)
        query = """
            SELECT "name", "adm_campus__r_name", "eventapi__start_date__c"
            FROM eventos
            WHERE ("adm_campus__r_name" ILIKE %s OR "name" ILIKE %s)
            AND "eventapi__start_date__c" >= CURRENT_DATE
            ORDER BY "eventapi__start_date__c" ASC
            LIMIT 3;
        """
        # Parametrización segura: evita SQL injection y permite wildcard
        parametro_busqueda = f"%{campus_detectado}%"
        cur.execute(query, (parametro_busqueda, parametro_busqueda))
        resultados = cur.fetchall()

        # Cierre de recursos
        cur.close()
        conn.close()

        if not resultados:
            return f"SIN_RESULTADOS: Se buscó en el campus '{campus_detectado}' pero no hay sesiones informativas futuras programadas en la base de datos."

        # 3) Formatear resultados en texto lineal para normalización posterior
        for evento in resultados:
            nombre_evento = evento[0]
            sede = evento[1]
            fecha = evento[2]
            info_eventos += f"- Evento: {nombre_evento} | Sede: {sede} | Fecha: {fecha}\n"

        # 4) Normalizador: limpia nombre y estandariza fechas
        prompt = ChatPromptTemplate.from_messages([
          ("system", """Eres un experto en normalización y limpieza de datos.
          Recibirás una lista de eventos. Tu trabajo es reescribir cada línea siguiendo estas REGLAS ESTRICTAS:

          1. **LIMPIEZA DEL NOMBRE (Campo Evento):**
            - Elimina cualquier fecha (dd/mm/aaaa, años 2024/2025, meses) que aparezca DENTRO del texto del nombre.
            - Elimina guiones, puntos o palabras conectoras ('del', 'el', 'en') si quedan colgadas al final del nombre tras borrar la fecha.
            - ¡IMPORTANTE!: La fecha escrita en el nombre suele ser ERRÓNEA o antigua. IGNORELA y BÓRRELA.

          2. **FECHA CORRECTA (Campo Fecha):**
            - ÚNICAMENTE confía en la fecha que aparece al final de la línea después de "| Fecha:".
            - Formatea esta fecha obligatoriamente así: **dd/mmm/aaaa** (ej: 07/feb/2026).
            - Usa las tres primeras letras del mes en minúsculas.

          3. **FORMATO DE SEDE:**
            - Asegura que la sede tenga un formato limpio y consistente.

          Ejemplo Entrada:
          - Evento: Sesión Informativa 15/Oct/2023 | Sede: Campus MTY | Fecha: 2026-10-15

          Ejemplo Salida:
          - Evento: Sesión Informativa | Sede: Campus MTY | Fecha: 15/oct/2026
          """),
          ("user", f"Limpia y normaliza este bloque de datos:\n\n{info_eventos}")
        ])

        chain = prompt | chat_model
        return chain.invoke({})

    except Exception as e:
        return f"ERROR_SISTEMA: Hubo un error al conectar con la base de datos de eventos: {str(e)}"

def obtener_tokens_langchain(response):
    """Extrae tokens desde el diccionario response_metadata del response de LangChain."""
    if(isinstance(response, str)):
        return 0, 0, response
    else:
        token_usage = response.response_metadata.get("token_usage", {})
        input_tokens = token_usage.get("prompt_tokens", 0)
        output_tokens = token_usage.get("completion_tokens", 0)

        return input_tokens, output_tokens, response.content.strip()

# ============================================================
# 7) Orquestador (Router)
# ============================================================
def ejecutar_langchain(user_input, retriever, judge_llm, model_key= 'gpt_5_chat'):
    """
    Orquestador/Router:
    - Clasifica la intención del usuario en: BECAS, CARRERAS o EVENTOS.
    - Ejecuta el especialista correspondiente.
    - Pasa la salida al sintetizador para generar la respuesta final.

    Patrón recomendado:
    - Separar "decisión" (routing) de "ejecución" (tools/RAG/DB) y de "presentación" (sintetizador).
    - Facilita evaluación comparativa: puedes medir accuracy del router,
      éxito por tarea y calidad de respuesta final por separado.
    """
    print("=" * 50)
    print("[ARCHITECTURE] Langchain")
    print(f"[MODEL] {model_key}")
    print("-" * 50)
    print(f"\n❓[LOG] Pregunta: {user_input}")

    # Registro de tiempo cuando inicia la función principal
    t_inicio_principal = datetime.now()
    chat_model = get_chat_model(model_key)
    tokens_in, tokens_out = 0, 0

    # Prompt de clasificación: salida restringida a 3 etiquetas (control del flujo)
    t_inicio_orquestador = datetime.now()
    clasificador_prompt = ChatPromptTemplate.from_messages([
        ("system", """Eres un coordinador de tareas. Tu único trabajo es clasificar la petición del usuario en una de estas tres categorías:
        1. BECAS: Si el usuario busca información sobre becas, apoyos económicos, requisitos o procesos de solicitud.
        2. CARRERAS: Si el usuario busca información sobre programas académicos, planes de estudio, carreras universitarias, cursos o información de admisión.
        3. EVENTOS: Si el usuario busca sesiones informativas de admisión específicas de algún campus.

        Responde ÚNICAMENTE con la palabra 'BECAS', 'CARRERAS' o 'EVENTOS' (sin puntos, sin explicación)."""),
        ("user", f"Clasifica esto: {user_input}")
    ])

    # Ejecutar decisión del router
    decision_chain = clasificador_prompt | chat_model

    resp_route = decision_chain.invoke({})

    t_fin_orquestador = datetime.now()

    # Acumular tokens del orquestador
    in_q, out_q, decision = obtener_tokens_langchain(resp_route)
    tokens_in += in_q
    tokens_out += out_q

    decision = decision.upper()

    print(f"🤖 [LOG] Orquestador redirigió a: {decision}")

    # Ejecutar especialista según clasificación
    t_inicio_especialista = datetime.now()

    respuesta_tecnica = ""
    if "BECAS" in decision:
        resp_tecnica = ejecutar_becas(user_input, chat_model)
    elif "CARRERAS" in decision:
        resp_tecnica = ejecutar_carreras(user_input, chat_model)
    elif "EVENTOS" in decision:
        resp_tecnica = ejecutar_eventos(user_input, chat_model)
    else:
        return "Lo siento, no pude identificar el área de tu consulta."

    # Acumular tokens del especialista
    in_s, out_s, respuesta_tecnica = obtener_tokens_langchain(resp_tecnica)
    tokens_in += in_s
    tokens_out += out_s

    t_fin_especialista = datetime.now()

    # Sintetizador
    #print(f"✍️ [LOG] Sintetizador generando respuesta final...\n")
    resp_final_obj = ejecutar_sintetizador(user_input, respuesta_tecnica, chat_model)

    # Acumular tokens del sintetizador
    in_s, out_s, respuesta_final = obtener_tokens_langchain(resp_final_obj)
    tokens_in += in_s
    tokens_out += out_s

    # Registro de tiempo cuando termina la función principal
    t_fin_principal = datetime.now()

    print(f"✅ [LOG] Respuesta:")
    print(respuesta_final)

    # Se ejecuta calculo de faithfulness
    score = evaluate_models(user_input, judge_llm, retriever, respuesta_final)

    # Registro de logs en el mismo DataFrame global
    global df_logs
    nuevo_registro = {
        "arquitectura": "Langchain",
        "modelo": model_key,
        "pregunta_usuario": user_input,
        "especialista": decision,
        "respuesta": respuesta_final,
        "inicio_principal": t_inicio_principal,
        "fin_principal": t_fin_principal,
        "inicio_especialista": t_inicio_especialista,
        "fin_especialista": t_fin_especialista,
        "inicio_orquestador": t_inicio_orquestador,
        "fin_orquestador": t_fin_orquestador,
        "total_tokens_entrada": tokens_in,
        "total_tokens_salida": tokens_out,
        "latencia_total": (t_fin_principal - t_inicio_principal).total_seconds(),
        "latencia_orquestador": (t_fin_orquestador - t_inicio_orquestador).total_seconds(),
        "latencia_especialista": (t_fin_especialista - t_inicio_especialista).total_seconds(),
        "overhead coordinacion": ((t_fin_principal - t_inicio_principal).total_seconds()) - ((t_fin_orquestador - t_inicio_orquestador).total_seconds()),
        "score": score
    }

    df_logs = pd.concat([df_logs, pd.DataFrame([nuevo_registro])], ignore_index=True)

    return respuesta_final
```

## 5.3. Microsoft Agent Framework

El Microsoft Agent Framework (basado principalmente en la biblioteca AutoGen) representa un paradigma avanzado en la computación agéntica, donde la unidad fundamental no es solo la "cadena" de pasos, sino la interacción conversacional entre múltiples entidades autónomas. A diferencia de otros enfoques, este framework facilita la creación de flujos de trabajo donde los agentes pueden dialogar entre sí para resolver tareas complejas de manera colaborativa. En esta implementación, se utiliza para gestionar la lógica de los especialistas de una forma ligera y modular, demostrando su eficiencia en la recuperación léxica y la síntesis de respuestas sin necesidad de orquestaciones secuenciales rígidas.

De acuerdo con la investigación fundamental de Microsoft sobre este ecosistema:

> "AutoGen permite el desarrollo de aplicaciones de LLM utilizando múltiples agentes que pueden conversar entre sí para resolver tareas. Los agentes de AutoGen son personalizables, conversables y permiten sin problemas la participación humana, operando en diversos modos que emplean combinaciones de LLMs, herramientas y humanos." (Wu et al., 2023).

```python
# =========================
# Imports / Dependencias
# =========================
import os
import re
import psycopg2

from agent_framework.azure import AzureOpenAIChatClient
from agent_framework.openai import OpenAIChatClient
from agent_framework import Agent

from functools import lru_cache

# Dependencias ligeras para RAG
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# KB local
FILE_KB_PATH = os.getenv("FILE_KB_PATH", "kb_total.md")

# =========================
# Model Registry (MAF)
# =========================

FOUNDRY_MODEL_REGISTRY = {
    "deepseek_v3_2": {
        "model_id": "DeepSeek-V3.2",
        "name": "DeepSeek-V3.2",
    },
    "gpt_5_chat": {
        "model_id": "gpt-5-chat",
        "name": "gpt-5-chat",
    },
    "grok_3": {
        "model_id": "grok-3",
        "name": "grok-3",
    },
    "llama_4_maverick": {
        "model_id": "Llama-4-Maverick-17B-128E-Instruct-FP8",
        "name": "Llama-4-Maverick-17B-128E-Instruct-FP8",
    },
}

# ============================================================
# 1) Inicialización del cliente (Chat) — Microsoft Agent Framework
# ============================================================

@lru_cache(maxsize=16)
def get_chat_client(model_key: str):
    """
    Crea un OpenAIChatClient apuntando al endpoint OpenAI/v1 de Azure Foundry.
    Cambiar de modelo = cambiar model_id (sin cambiar código del pipeline).
    """
    if model_key not in FOUNDRY_MODEL_REGISTRY:
        raise ValueError(
            f"Modelo '{model_key}' no existe. Opciones: {list(FOUNDRY_MODEL_REGISTRY.keys())}"
        )

    base_url = os.getenv("FOUNDRY_OPENAI_BASE_URL")
    api_key  = os.getenv("FOUNDRY_OPENAI_API_KEY")

    if not base_url or not api_key:
        raise ValueError("Faltan variables: FOUNDRY_OPENAI_BASE_URL y/o FOUNDRY_OPENAI_API_KEY")

    cfg = FOUNDRY_MODEL_REGISTRY[model_key]

    # Cliente OpenAI-compatible (v1) para Foundry Models
    # En v1 se evita api-version y se usa /openai/v1/.  [oai_citation:3‡Microsoft Learn](https://learn.microsoft.com/es-es/azure/ai-foundry/openai/api-version-lifecycle?view=foundry-classic&utm_source=chatgpt.com)
    return OpenAIChatClient(
        base_url=base_url,
        api_key=api_key,
        model_id=cfg["model_id"],
    )

# ============================================================
# 2) RAG local (TF-IDF) para Becas y Carreras
# ============================================================
# Diseño:
# - Leemos el Markdown, lo “troceamos” en chunks razonables
# - Construimos un índice TF-IDF una sola vez (cache en memoria)
# - Exponemos tools que devuelven:
#   - contexto concatenado
#   - “citas” simples (chunk_id + preview)
#
# Importante:
# - Este RAG NO usa embeddings/vectores semánticos; usa recuperación léxica (TF-IDF).
# - Sigue siendo RAG porque “aumenta” el prompt con contexto recuperado.
# - Para migrar a vector search, reemplaza el motor de búsqueda en las tools.
# ============================================================
TOOL_CALLS = {"becas": 0, "carreras": 0}

def _read_markdown(path: str) -> str:
    """Carga el archivo Markdown (KB)."""
    with open(path, "r", encoding="utf-8") as f:
        return f.read()

def _normalize_text(text: str) -> str:
    """Normaliza texto para mejorar recall en TF-IDF."""
    text = text.lower()
    text = re.sub(r"\s+", " ", text).strip()
    return text

def _chunk_markdown(md: str, chunk_size: int = 900, overlap: int = 150) -> list[str]:
    """
    Divide el markdown en chunks “razonables”.
    Estrategia:
    - Primero separa por encabezados y saltos grandes.
    - Luego hace “ventanas” con overlap para no cortar ideas importantes.
    """
    # Separación inicial por encabezados/líneas vacías grandes
    blocks = re.split(r"\n(?=#+\s)|\n{2,}", md)

    # Limpieza de bloques vacíos
    blocks = [b.strip() for b in blocks if b and b.strip()]

    # Construcción de chunks con tamaño objetivo + overlap
    chunks: list[str] = []
    buffer = ""

    for b in blocks:
        if len(buffer) + len(b) + 1 <= chunk_size:
            buffer = (buffer + "\n" + b).strip()
        else:
            if buffer:
                chunks.append(buffer)
            buffer = b

    if buffer:
        chunks.append(buffer)

    # Overlap (a nivel caracteres) para no perder continuidad
    if overlap > 0 and len(chunks) > 1:
        overlapped = []
        for i, ch in enumerate(chunks):
            if i == 0:
                overlapped.append(ch)
            else:
                prev = chunks[i - 1]
                tail = prev[-overlap:] if len(prev) > overlap else prev
                overlapped.append((tail + "\n" + ch).strip())
        chunks = overlapped

    return chunks

class LocalTfidfRAGIndex:
    """
    Índice RAG local (TF-IDF).
    - Construye vectorizador + matriz TF-IDF para los chunks.
    - Permite búsqueda top-k por similitud coseno.
    """

    def __init__(self, chunks: list[str]):
        self.chunks = chunks
        self.norm_chunks = [_normalize_text(c) for c in chunks]

        # Vectorizador TF-IDF (ligero y rápido)
        self.vectorizer = TfidfVectorizer(
            stop_words=None,     # español + dominio; preferimos no filtrar agresivo
            ngram_range=(1, 2),   # bigramas suelen ayudar en KB institucional
            min_df=1,
        )

        # Matriz documento-término
        self.tfidf_matrix = self.vectorizer.fit_transform(self.norm_chunks)

    def search(self, query: str, top_k: int = 4) -> list[tuple[int, float, str]]:
        """
        Retorna lista de (chunk_id, score, chunk_text).
        """
        q = _normalize_text(query)
        q_vec = self.vectorizer.transform([q])
        sims = cosine_similarity(q_vec, self.tfidf_matrix).flatten()

        if sims.size == 0:
            return []

        top_idx = np.argsort(sims)[::-1][:top_k]
        results = [(int(i), float(sims[i]), self.chunks[int(i)]) for i in top_idx if sims[i] > 0.0]
        return results

# --- Construcción del índice (con tolerancia a fallos) ---
# Mantener índices separados permite ajustar reglas/queries por dominio sin mezclar.
try:
    kb_md = _read_markdown(FILE_KB_PATH)
    kb_chunks = _chunk_markdown(kb_md, chunk_size=900, overlap=150)

    rag_index_becas = LocalTfidfRAGIndex(kb_chunks)
    rag_index_carreras = LocalTfidfRAGIndex(kb_chunks)

    print(f"✅ KB cargada e indexada: {len(kb_chunks)} chunks (TF-IDF).")
except Exception as e:
    rag_index_becas = None
    rag_index_carreras = None
    print(f"⚠️ No se pudo construir el índice RAG local desde '{FILE_KB_PATH}': {e}")

def _format_rag_results(results: list[tuple[int, float, str]]) -> str:
    """
    Formatea los resultados recuperados para inyectarlos como CONTEXTO.
    Incluye una “cita” simple por chunk (ID + snippet).
    """
    lines = []
    for chunk_id, score, text in results:
        snippet = text[:220].replace("\n", " ").strip()
        lines.append(
            f"[KB_CHUNK:{chunk_id} | score:{score:.3f}] {snippet}..."
        )
    return "\n".join(lines)

def buscar_becas_kb(pregunta: str, top_k: int = 4) -> str:
    TOOL_CALLS["becas"] += 1
    #print(f"[TOOL] buscar_becas_kb called ({TOOL_CALLS['becas']}) | top_k={top_k}", flush=True)

    if rag_index_becas is None:
        return "FALTA_INFORMACION: No está disponible el índice local de la KB (kb_total.md)."

    query = f"becas apoyos económicos financiamiento requisitos proceso: {pregunta}"
    results = rag_index_becas.search(query, top_k=top_k)

    # Print corto: solo cantidad (para no saturar)
    #print(f"[TOOL] becas hits={len(results)}", flush=True)

    if not results:
        return "SIN_RESULTADOS: No encontré snippets relevantes en la KB para becas."

    return "## CONTEXTO RECUPERADO (KB - Becas)\n" + _format_rag_results(results)

def buscar_carreras_kb(pregunta: str, top_k: int = 4) -> str:
    TOOL_CALLS["carreras"] += 1
    #print(f"[TOOL] buscar_carreras_kb called ({TOOL_CALLS['carreras']}) | top_k={top_k}", flush=True)

    if rag_index_carreras is None:
        return "FALTA_INFORMACION: No está disponible el índice local de la KB (kb_total.md)."

    query = f"carreras programas académicos plan de estudios materias enfoque: {pregunta}"
    results = rag_index_carreras.search(query, top_k=top_k)

    #print(f"[TOOL] carreras hits={len(results)}", flush=True)

    if not results:
        return "SIN_RESULTADOS: No encontré snippets relevantes en la KB para carreras."

    return "## CONTEXTO RECUPERADO (KB - Carreras)\n" + _format_rag_results(results)

# ============================================================
# 4) Tool de Base de Datos (Eventos)
# ============================================================
def buscar_eventos_db(campus: str):
    """
    Tool: consulta SQL a PostgreSQL para buscar eventos próximos en un campus.

    Buenas prácticas:
    - Usar parámetros (evita SQL injection)
    - Cerrar cursor/conn
    - Retornar strings “técnicas” simples que luego el orquestador “humaniza”
    """
    try:
        conn = psycopg2.connect(
            host=BD_HOST,
            database=BD_SCHEMA,
            user=BD_USER,
            password=BD_PASSWORD,
            port=5432
        )
        cur = conn.cursor()
        query = """
            SELECT "name", "adm_campus__r_name", "eventapi__start_date__c"
            FROM eventos
            WHERE ("adm_campus__r_name" ILIKE %s OR "name" ILIKE %s)
            AND "eventapi__start_date__c" >= CURRENT_DATE
            ORDER BY "eventapi__start_date__c" ASC
            LIMIT 3;
        """
        cur.execute(query, (f"%{campus}%",f"%{campus}%"))
        resultados = cur.fetchall()
        cur.close()
        conn.close()

        if not resultados:
            return "SIN_RESULTADOS: No hay eventos programados para este campus."

        return "\n".join([f"Evento: {r[0]} | Sede: {r[1]} | Fecha: {r[2]}" for r in resultados])
    except Exception as e:
        return f"ERROR_SISTEMA: Error en DB: {str(e)}"

# ============================================================
# 5) Agentes especializados (Becas / Carreras / Eventos)
# ============================================================
# Importante:
# - Para RAG: registramos las tools buscar_*_kb en sus agentes.
# - El agente DEBE:
#   1) llamar la tool
#   2) responder SOLO con lo que aparezca en el contexto recuperado
#   3) si no hay contexto suficiente -> FALTA_INFORMACION
# ============================================================


# Agente encargado de temas de financiamiento y becas
# =========================
# 1) Agentes (MAF)
# =========================
def build_agents(chat_client):
    agente_becas = chat_client.as_agent(
        name="Agente_Becas",
        instructions="""
    Eres especialista en BECAS del Tec de Monterrey.

    REGLAS:
    - Responde SOLO con la información contenida en el CONTEXTO.
    - Si el CONTEXTO dice SIN_RESULTADOS o FALTA_INFORMACION, responde iniciando con "FALTA_INFORMACION:" y pide el dato faltante.
    - No inventes montos, porcentajes, fechas ni requisitos.
    - Respuesta técnica (sin tono marketing), clara y estructurada.
    """.strip()
    )

    agente_carreras = chat_client.as_agent(
        name="Agente_Carreras",
        instructions="""
    Eres especialista en CARRERAS/PROGRAMAS ACADÉMICOS del Tec de Monterrey.

    REGLAS:
    - Responde SOLO con la información contenida en el CONTEXTO.
    - Si el CONTEXTO dice SIN_RESULTADOS o FALTA_INFORMACION, responde iniciando con "FALTA_INFORMACION:" y pide el dato faltante.
    - No inventes plan de estudios, duración, costos, sedes o fechas.
    - Respuesta técnica, clara y estructurada.
    """.strip()
    )

    agente_eventos = chat_client.as_agent(
        name="Agente_Eventos",
        instructions="""
    Eres un experto en normalización y limpieza de datos.
    Recibirás una lista de eventos. Tu trabajo es reescribir cada línea siguiendo estas REGLAS ESTRICTAS:

    1. **LIMPIEZA DEL NOMBRE (Campo Evento):**
      - Elimina cualquier fecha (dd/mm/aaaa, años 2024/2025, meses) que aparezca DENTRO del texto del nombre.
      - Elimina guiones, puntos o palabras conectoras ('del', 'el', 'en') si quedan colgadas al final del nombre tras borrar la fecha.
      - ¡IMPORTANTE!: La fecha escrita en el nombre suele ser ERRÓNEA o antigua. IGNORELA y BÓRRELA.

    2. **FECHA CORRECTA (Campo Fecha):**
      - ÚNICAMENTE confía en la fecha que aparece al final de la línea después de "| Fecha:".
      - Formatea esta fecha obligatoriamente así: **dd/mmm/aaaa** (ej: 07/feb/2026).
      - Usa las tres primeras letras del mes en minúsculas.

    3. **FORMATO DE SEDE:**
      - Asegura que la sede tenga un formato limpio y consistente.

    Ejemplo Entrada:
    - Evento: Sesión Informativa 15/Oct/2023 | Sede: Campus MTY | Fecha: 2026-10-15

    Ejemplo Salida:
    - Evento: Sesión Informativa | Sede: Campus MTY | Fecha: 15/oct/2026
    """.strip()
    )

    sintetizador = chat_client.as_agent(
        name="Sintetizador",
        instructions="""
    Eres un sintetizador de información experto. Actúa como un asistente de atención al usuario **amable, empático y muy eficiente**.

    Tu tarea es comunicar la información técnica al usuario siguiendo estas REGLAS DE ORO:

    1. **BREVEDAD RADICAL:** Tu respuesta debe ser breve pero no descartes información importante para el usuario. Sé directo pero suave.
    2. **TONO:** Usa un lenguaje cercano y servicial (Ej: "Con gusto te comparto...", "Claro, aquí tienes...").
    3. **COHERENCIA:** Responde directamente a la pregunta del usuario usando SOLO la información provista.
    4. **SIN RELLENO:** Elimina saludos largos o despedidas innecesarias. Ve al grano con amabilidad.
    5. NO inventes información extra (no alucinar).
    6. NO adivines fechas o requisitos que no estén en la respuesta técnica.
    7. Si la respuesta técnica dice que falta información, menciónalo tal cual, no menciones tecnicismos acerca de bases de datos, conexiones o errores.
    8. Asegúrate de responder directamente a lo que el usuario preguntó originalmente.
    9. Si se trata de eventos, muestra una lista completa de las opciones con un formato entendible y amigable para el usuario.
    """.strip()
    )

    # ============================================================
    # 6) Orquestador (Router + síntesis)
    # ============================================================
    # Este agente:
    # 1) Decide si la consulta es BECAS / CARRERAS / EVENTOS
    # 2) Delega al especialista correspondiente (usando agents=[...])
    # 3) Devuelve una respuesta final breve y amable
    # ============================================================
    orquestador = chat_client.as_agent(
        name="Orquestador_Tec",
        instructions="""
    Eres un coordinador de tareas. Tu único trabajo es clasificar la petición del usuario en una de estas tres categorías:
        1. BECAS: Si el usuario busca información sobre becas, apoyos económicos, requisitos o procesos de solicitud.
        2. CARRERAS: Si el usuario busca información sobre programas académicos, planes de estudio, carreras universitarias, cursos o información de admisión.
        3. EVENTOS: Si el usuario busca sesiones informativas de admisión específicas de algún campus.

    Responde ÚNICAMENTE con la palabra 'BECAS', 'CARRERAS' o 'EVENTOS' (sin puntos, sin explicación).
    """.strip()
    )

    return orquestador, agente_becas, agente_carreras, agente_eventos, sintetizador

def _normalize_decision(text: str) -> str:
    t = (text or "").strip().upper()
    # tolerancia por si el LLM agrega algo
    if "BECAS" in t:
        return "BECAS"
    if "CARRERAS" in t:
        return "CARRERAS"
    if "EVENTOS" in t:
        return "EVENTOS"
    return "UNKNOWN"


def _build_context_for(decision: str, pregunta_usuario: str) -> str:
    """
    Construye el CONTEXTO que se le pasa al especialista.
    Aquí garantizas que sí se usa KB/DB, sin depender de tools del framework.
    """
    if decision == "BECAS":
        return buscar_becas_kb(pregunta_usuario)

    if decision == "CARRERAS":
        return buscar_carreras_kb(pregunta_usuario)

    if decision == "EVENTOS":
        # aquí puedes usar tu extractor de campus o pedirlo si falta
        campus = None
        m = re.search(r"\ben\s+([A-Za-zÁÉÍÓÚÜÑáéíóúüñ\s]+)\??$", pregunta_usuario.strip())
        if m:
            campus = m.group(1).strip()

        if not campus:
            return "FALTA_INFORMACION: Falta campus/ciudad para buscar eventos. Ejemplos: Monterrey, Guadalajara, Querétaro."
        return buscar_eventos_db(campus)

    return "FALTA_INFORMACION: No se pudo clasificar la consulta."

def obtener_tokens(response):
    """Extrae el total_tokens de usage_details de forma segura."""
    # Intentar obtener el atributo
    details = getattr(response, 'usage_details', None)
    if details:
        # Algunos modelos devuelven un objeto con .total_tokens
        if hasattr(details, 'total_tokens'):
            return details.total_tokens
        # Otros devuelven un diccionario
        if isinstance(details, dict):
            return details.get('total_tokens', 0)

    # Si sigue fallando, buscamos en el último mensaje del historial si existe
    if hasattr(response, 'messages') and response.messages:
        last_msg = response.messages[-1]
        msg_details = getattr(last_msg, 'usage_details', None)
        if msg_details:
            return getattr(msg_details, 'total_tokens', 0)

    return 0

# ============================================================
# 7) Runner
# ============================================================
async def ejecutar_microsoft_agent_framework(pregunta_usuario: str, retriever, judge_llm, model_key: str = 'gpt_5_chat'):
    # Registro de tiempo cuando inicia la función principal
    t_inicio_principal = datetime.now()

    chat_client = get_chat_client(model_key)
    orquestador, agente_becas, agente_carreras, agente_eventos, sintetizador = build_agents(chat_client)
    print("=" * 50)
    print("[ARCHITECTURE] Microsoft Agent Framework")
    print(f"[MODEL] {model_key}")
    print("-" * 50)
    print(f"\n❓[LOG] Pregunta: {pregunta_usuario}")

    # Registro de tiempo cuando termina el orquestador
    t_inicio_orquestador = datetime.now()

    # (1) Orquestador: clasifica
    resp_route = await orquestador.run(pregunta_usuario)

    # Registro de tiempo cuando termina el orquestador
    t_fin_orquestador = datetime.now()

    decision_raw = resp_route.text if hasattr(resp_route, "text") else str(resp_route)
    decision = _normalize_decision(decision_raw)

    print(f"🤖 [LOG] Orquestador redirigió a: {decision}")

    # (2) Contexto (RAG/DB) determinístico
    contexto = _build_context_for(decision, pregunta_usuario)

    #print(f"\n[CTX] CONTEXTO RECUPERADO ({decision}):\n")
    #print(contexto)
    #print("\n" + "-" * 50)

    # (3) Especialista: SOLO el agente elegido
    if decision == "BECAS":
        especialista = agente_becas
    elif decision == "CARRERAS":
        especialista = agente_carreras
    elif decision == "EVENTOS":
        especialista = agente_eventos
    else:
        # fallback seguro
        especialista = agente_becas

    prompt_especialista = f"""
PREGUNTA_USUARIO:
{pregunta_usuario}

CONTEXTO:
{contexto}

RESPUESTA_TECNICA:
""".strip()

    # Registro de tiempo cuando inicia uno de los especialistas
    t_inicio_especialista = datetime.now()

    resp_tec = await especialista.run(prompt_especialista)

    # Registro de tiempo cuando termina el especialista
    t_fin_especialista = datetime.now()

    respuesta_tecnica = resp_tec.text if hasattr(resp_tec, "text") else str(resp_tec)

    # print mínimo para debug
    #print(f"[TECH] {especialista.name} respondió ({len(respuesta_tecnica)} chars)", flush=True)

    # (4) Sintetizador: respuesta final
    prompt_sintetizador = f"""
PREGUNTA_USUARIO:
{pregunta_usuario}

RESPUESTA_TECNICA:
{respuesta_tecnica}

RESPUESTA_FINAL:
""".strip()

    resp_final = await sintetizador.run(prompt_sintetizador)

    # Registro de tiempo cuando termina la función principal
    t_fin_principal = datetime.now()

    respuesta_final = resp_final.text if hasattr(resp_final, "text") else str(resp_final)

    print("✅ [LOG]  Respuesta:")
    print(respuesta_final)

    # Se ejecuta calculo de faithfulness
    score = evaluate_models(pregunta_usuario, judge_llm, retriever, respuesta_final)

    # Registro del consumo total de tokens
    total_tokens_entrada = (
        resp_route.usage_details["input_token_count"] +
        resp_tec.usage_details["input_token_count"] +
        resp_final.usage_details["input_token_count"]
    )

    total_tokens_salida = (
        resp_route.usage_details["output_token_count"] +
        resp_tec.usage_details["output_token_count"] +
        resp_final.usage_details["output_token_count"]
    )

    # Registro de logs
    global df_logs
    nuevo_registro = {
        "arquitectura": "Microsoft Agent Framework",
        "modelo": model_key,
        "pregunta_usuario": pregunta_usuario,
        "especialista": decision,
        "respuesta": respuesta_final,
        "inicio_principal": t_inicio_principal,
        "fin_principal": t_fin_principal,
        "inicio_especialista": t_inicio_especialista,
        "fin_especialista": t_fin_especialista,
        "inicio_orquestador": t_inicio_orquestador,
        "fin_orquestador": t_fin_orquestador,
        "total_tokens_entrada": total_tokens_entrada,
        "total_tokens_salida": total_tokens_salida,
        "latencia_total": (t_fin_principal - t_inicio_principal).total_seconds(),
        "latencia_orquestador": (t_fin_orquestador - t_inicio_orquestador).total_seconds(),
        "latencia_especialista": (t_fin_especialista - t_inicio_especialista).total_seconds(),
        "overhead coordinacion": ((t_fin_principal - t_inicio_principal).total_seconds()) - ((t_fin_orquestador - t_inicio_orquestador).total_seconds()),
        "score": score
    }

    df_logs = pd.concat([df_logs, pd.DataFrame([nuevo_registro])], ignore_index=True)

    return respuesta_final
```

El experimento se estructuró como una prueba matricial de tres dimensiones: Preguntas × Modelos × Frameworks. Este enfoque asegura que cada variable sea probada en igualdad de condiciones, permitiendo un análisis comparativo estadísticamente válido.

Al utilizar consultas reales, el experimento garantiza una alta validez externa.

**2. Variables de Control**

Como señalan Gun y Oksanen (2025), la comparación entre diferentes configuraciones de modelos es esencial para identificar cuál gestiona mejor las tareas de respuesta a preguntas técnicas en entornos de recuperación.

Frameworks: La comparación se centró en las implementaciones de LangChain y Microsoft Agent Framework (MAF).

**3. Metodología de Ejecución**

El proceso se automatizó mediante una función asíncrona que itera sobre cada combinación posible. Por cada iteración (pregunta-modelo-arquitectura), el sistema. De acuerdo con Singh et al. (2025), este tipo de evaluación agéntica es fundamental para medir cómo los agentes autónomos integrados en el pipeline de RAG gestionan tareas que requieren un razonamiento complejo y multietapa.

1. Activa el Orquestador correspondiente.

2. Ejecuta el flujo de recuperación y síntesis.

3. Registro de Logs: Captura eventos críticos en un registro detallado que incluye tiempos de respuesta, precisión de la recuperación y coherencia de la salida.

Todos estos eventos se consolidan en un DataFrame de Pandas (df_logs), el cual actúa como la fuente de verdad para el posterior análisis de métricas y generación de resultados.

```python
import time
import asyncio
import nest_asyncio

nest_asyncio.apply()

async def ejecutar_experimentos():
    df_logs = pd.DataFrame()

    preguntas = [
        "¿Cuáles son los requisitos para solicitar una beca al Talento Emprendedor?",
        "¿Cuáles son los requisitos para ser admitido? (Si provengo de Prepa Tec)",
        "¿Cuáles son los requisitos para ser admitido? (Si provengo de prepa No Tec)",
        "¿Cuáles son los requisitos para solicitar una beca al Talento por Liderazgo?",
        "¿Cómo puedo saber más sobre qué carrera elegir si quiero estudiar en campus Monterrey?",
        "¿Cuáles son los requisitos para solicitar una beca al Talento Atlético?",
        "¿A qué becas puedo aplicar?",
        "¿Por qué estudiar Ambiente Construido?",
        "¿Cuáles son los requisitos para solicitar una beca Socioeconómica?",
        "¿Qué son las sesiones informativas?",
        "¿Cuáles son los requisitos para solicitar una beca Líderes del mañana?",
        "¿Cuántos alumnos de profesional hay en Campus San Luis Potosí?",
        "¿Cómo puedo saber más del campus Saltillo?",
        "¿Cuál es la mínima carga que debo llevar para tener beca?",
        "¿Por qué debería estudiar en el Tec de Monterrey?",
        "¿En qué puedo especializarme si estudio IDS Ingeniero en Desarrollo Sustentable?",
        "¿Cómo puedo saber más sobre la oferta educativa de campus guadalajara?",
        "¿Cuál es el campo laboral  para la carrera de Licenciado en Animación y Arte Digital?",
        "¿Cuál es el campo laboral  para la carrera de Arquitectura?",
        "¿Qué es el semestre Tec?",
        "¿Por qué estudiar Ingeniería Innovación y Transformación?",
    ]

    modelos = [
        "gpt_5_chat",
        "grok_3",
        "llama_4_maverick",
        "deepseek_v3_2",
    ]

    arquitecturas = [
        "Langchain",
        "Microsoft Agent Framework"
    ]

    # Retriever con base de conocimiento
    retriever = _build_kb_retriever_from_md()

    # Estructura simplificada del flujo experimental
    for arquitectura in arquitecturas:
        for pregunta in preguntas:
            for modelo in modelos:
                # Ejecución dinámica según el framework seleccionado
                print('Pregunta #', preguntas.index(pregunta) + 1)

                #Modelo juez evaluador
                judge_llm = get_chat_model(modelo)

                await asyncio.sleep(10)
                if arquitectura == "Langchain":
                    respuesta = ejecutar_langchain(pregunta, retriever, judge_llm, modelo)
                else:
                    respuesta = await ejecutar_microsoft_agent_framework(pregunta, retriever, judge_llm, modelo)

                print('\n\n')

await ejecutar_experimentos()
```
